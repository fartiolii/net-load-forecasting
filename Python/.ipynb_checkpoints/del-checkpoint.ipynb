## Manually searching for theta

```python
def features_matrix_standardized(GAM_model, df):
    n_features = df.shape[1]+1
    features_eval = np.zeros((df.shape[0],n_features))
    it = 0
    for f in range(n_features):
        feature_mat = GAM_model._modelmat(np.array(df), term=f)
        n_coeff_feature = feature_mat.shape[1]
        features_eval[:, f] = feature_mat@GAM_model.coef_[it:it+n_coeff_feature]
        it += n_coeff_feature
    features_eval_std = (features_eval - features_eval.mean(axis=0))/features_eval.std(axis=0)
    return features_eval, features_eval_std
    
def features_matrix_update(GAM_model, df, features_eval, features_eval_std):
    n_features = df.shape[1]
    feat = np.zeros(n_features,)
    it = 0
    for f in range(n_features):
        feature_mat = GAM_model._modelmat(np.array(df.iloc[-1,:]).reshape(1,n_features), term=f)
        n_coeff_feature = feature_mat.shape[1]
        feat[f] = np.sum(feature_mat*GAM_model.coef_[it:it+n_coeff_feature])
        it += n_coeff_feature
    features_eval = np.r_[features_eval, feat.reshape(1,n_features)]
    features_eval_std = (features_eval - features_eval.mean(axis=0))/features_eval.std(axis=0)
    return features_eval, features_eval_std
    

def loss_fun_static_Kalman(theta,y,features):
    return np.sum((y-features@theta)**2)+np.linalg.norm(theta)

    
def theta_optimize(y,features,theta_0):
    res = minimize(loss_fun_static_Kalman, theta_0, args=(y,features))
    return res.x
```

```python
n_features = train_dataset_GroupC.shape[1]
features_eval_train, features_eval_train_std = features_matrix_standardized(gam_model, train_dataset_GroupC)
Yt_train = GroupC_df_GAM['node_n']
theta_0 = np.zeros(train_dataset_GroupC.shape[1],)
theta_opt = theta_optimize(Yt_train,features_eval_train_std,theta_0)
print("Parameters: ", theta_opt)
print("RMSE GAM: ", RMSE(Yt_train, features_eval_train@np.ones((n_features,))))
print("RMSE Static Kalman: ", RMSE(Yt_train, features_eval_train_std@theta_opt))
```

## Quantile Regression

```python
from scipy.optimize import minimize

def rho(y, yhat, q):
    return (1*(y<yhat)-q)*(yhat-y)

def standardize(df):
    return (df-df.mean())/df.std()

def zeta_vec(yhat, df):
    # function that returns the feature dataframe for quantile regression
    df['S_CS'] = df['SSRD_weighted.mean_cell']*df['SolarCap']
    features_qr = np.c_[np.ones(len(yhat),), yhat, yhat**2, df['S_CS'], df['WindSpd10_weighted.mean_cell'],
                    df['TP_weighted.mean_pcell'], 1*(df['dow_Rph']=='Lun'),1*(df['dow_Rph']=='Mar'),1*(df['dow_Rph']=='Mer'),
                    1*(df['dow_Rph']=='Jeu'),1*(df['dow_Rph']=='Ven'),1*(df['dow_Rph']=='Sam'),1*(df['dow_Rph']=='Dim'),1*(df['dow_Rph']=='Hol'),    
                    1*(df['clock_hour']>=0)*(df['clock_hour']<1), 1*(df['clock_hour']>=1)*(df['clock_hour']<2),
                    1*(df['clock_hour']>=2)*(df['clock_hour']<3), 1*(df['clock_hour']>=3)*(df['clock_hour']<4),
                    1*(df['clock_hour']>=4)*(df['clock_hour']<5), 1*(df['clock_hour']>=5)*(df['clock_hour']<6),
                    1*(df['clock_hour']>=6)*(df['clock_hour']<7), 1*(df['clock_hour']>=7)*(df['clock_hour']<8),
                    1*(df['clock_hour']>=8)*(df['clock_hour']<9), 1*(df['clock_hour']>=9)*(df['clock_hour']<10),
                    1*(df['clock_hour']>=10)*(df['clock_hour']<11), 1*(df['clock_hour']>=11)*(df['clock_hour']<12), 
                    1*(df['clock_hour']>=12)*(df['clock_hour']<13), 1*(df['clock_hour']>=13)*(df['clock_hour']<14),
                    1*(df['clock_hour']>=14)*(df['clock_hour']<15), 1*(df['clock_hour']>=15)*(df['clock_hour']<16),
                    1*(df['clock_hour']>=16)*(df['clock_hour']<17), 1*(df['clock_hour']>=17)*(df['clock_hour']<18),
                    1*(df['clock_hour']>=18)*(df['clock_hour']<19), 1*(df['clock_hour']>=19)*(df['clock_hour']<20), 
                    1*(df['clock_hour']>=20)*(df['clock_hour']<21), 1*(df['clock_hour']>=21)*(df['clock_hour']<22),
                    1*(df['clock_hour']>=22)*(df['clock_hour']<23), 1*(df['clock_hour']>=23)] 
                        
    return features_qr

def loss_fun_qr(beta,y,yhat,z,q):
    return np.sum(rho(y-yhat,z@beta,q))
    
def beta_opt(y,yhat,z,q,beta_0):
    res = minimize(loss_fun_qr, beta_0, args=(y,yhat,z,q))
    return res.x

def quantile(y,yhat,z,q,beta_0):
    beta = beta_opt(y,yhat,z,q,beta_0)
    return yhat + z@beta, beta

def pinball_score_alpha(estimated_quantile,y,alpha):
    return np.sum((estimated_quantile - y)*(1*(y <= estimated_quantile)-alpha))/len(y)

def pinball_score(estimated_quantile_mat, y, alpha_vec):
    score = 0
    for i in range(alpha_vec.shape[0]):
        score += pinball_score_alpha(estimated_quantile_mat[:,i],y,alpha_vec[i])
    return score/alpha_vec.shape[0]
```

```python
q_L = 0.025
q_U = 0.975

# Let's test this on the training set
yhat = y_train
y = GroupC_df_GAM['node']
df = GroupC_df_GAM.copy()
z = zeta_vec(yhat, df)
beta_0 = np.zeros(z.shape[1])

upper_quantile, beta_upper = quantile(y,yhat,z,q_U,beta_0)
lower_quantile, beta_lower = quantile(y,yhat,z,q_L,beta_0)

print("Pinball score upper quantile: ", pinball_score_alpha(upper_quantile,y,q_U))
print("Pinball score lower quantile: ", pinball_score_alpha(lower_quantile,y,q_L))
print("Pinball score combined quantiles: ", pinball_score(np.c_[upper_quantile,lower_quantile], y, np.array([q_U,q_L])))

# NB: prediction data is not standardized
```

# Kalman Filter

## Static Case

```python
def features_matrix_standardized(GAM_model, df):
    n_features = df.shape[1]
    features_eval = np.zeros(df.shape)
    it = 0
    for f in range(n_features):
        feature_mat = GAM_model._modelmat(np.array(df), term=f)
        n_coeff_feature = feature_mat.shape[1]
        features_eval[:, f] = feature_mat@GAM_model.coef_[it:it+n_coeff_feature]
        it += n_coeff_feature
    features_eval_std = (features_eval - features_eval.mean(axis=0))/features_eval.std(axis=0)
    #features_eval_train_std = np.c_[features_eval_train_std, np.ones((train_dataset_GroupC.shape[0],))]
    return features_eval, features_eval_std
    
def features_matrix_update(GAM_model, df, features_eval, features_eval_std):
    n_features = df.shape[1]
    feat = np.zeros(n_features,)
    it = 0
    for f in range(n_features):
        feature_mat = GAM_model._modelmat(np.array(df.iloc[-1,:]).reshape(1,n_features), term=f)
        n_coeff_feature = feature_mat.shape[1]
        feat[f] = np.sum(feature_mat*GAM_model.coef_[it:it+n_coeff_feature])
        it += n_coeff_feature
    features_eval = np.r_[features_eval, feat.reshape(1,n_features)]
    features_eval_std = (features_eval - features_eval.mean(axis=0))/features_eval.std(axis=0)
    return features_eval, features_eval_std
    

def loss_fun_static_Kalman(theta,y,features):
    return np.sum((y-features@theta)**2)+np.linalg.norm(theta)
    
def theta_optimize(y,features,theta_0):
    res = minimize(loss_fun_static_Kalman, theta_0, args=(y,features))
    return res.x
```

```python
# We obtain the evaluation of the features vector on the training set and we standardize it
n_features = train_dataset_GroupC.shape[1]
features_eval_train = np.zeros(train_dataset_GroupC.shape)
it = 0
for f in range(n_features):
    feature_mat = gam_model._modelmat(np.array(train_dataset_GroupC), term=f)
    n_coeff_feature = feature_mat.shape[1]
    features_eval_train[:, f] = feature_mat@gam_model.coef_[it:it+n_coeff_feature]
    it += n_coeff_feature
    
features_eval_train_std = (features_eval_train - features_eval_train.mean(axis=0))/features_eval_train.std(axis=0)
```

## We look for theta vector which minimizes the error on the training set 

```python
n_features = train_dataset_GroupC.shape[1]
features_eval_train, features_eval_train_std = features_matrix_standardized(gam_model, train_dataset_GroupC)
Yt_train = GroupC_df_GAM['node_n']
theta_0 = np.zeros(train_dataset_GroupC.shape[1],)
theta_opt = theta_optimize(Yt_train,features_eval_train_std,theta_0)
print("Parameters: ", theta_opt)
print("RMSE GAM: ", RMSE(Yt_train, features_eval_train@np.ones((n_features,))))
print("RMSE Static Kalman: ", RMSE(Yt_train, features_eval_train_std@theta_opt))
```

# Test set Evaluation

### Evaluation at each time t: we use data up to t-1 to compute the optimal theta for t

```python
theta_mat = np.zeros(test_dataset_GroupC.shape)

for t in range(len(test_dataset_GroupC)):
    updated_dataset = pd.concat([train_dataset_GroupC,test_dataset_GroupC.iloc[:t,:]])
    updated_Y = pd.concat([GroupC_df_GAM['node_n'],  GroupC_testSet['node_n'][:t]])
    if t==0:
        features_ev, features_ev_std = features_matrix_standardized(gam_model, updated_dataset)
        theta_0 = theta_mat[t,:]
    else:
        features_ev, features_ev_std = features_matrix_update(gam_model, updated_dataset, features_ev, features_ev_std)
        theta_0 = theta_mat[t-1,:]   
    if t%100==0:
        print(t)
    theta_mat[t,:] = theta_optimize(updated_Y,features_ev_std,theta_0)
```

## 2) Partial Auto Correlation Function PACF

```python
def Seasonality_Components_Solar(t, dow, doy, clock_hour, datetime):
    # seasonality on yearly scale
    wy = 2*np.pi/365
    month = datetime.dt.month
    is_summer = (month > 5)*(month < 9)
    # seasonality on weekly scale: we look for weekend days
    is_saturday = dow == 'Saturday'
    is_sunday = dow == 'Sunday'
    day_of_week = 1*(dow=='Monday')+2*(dow=='Tuesday')+3*(dow=='Wednesday')+4*(dow=='Thursday')+5*(dow=='Friday') \
                +6*(dow=='Saturday')+7*(dow=='Sunday')
    ww = 2*np.pi/7
    # seasonality on daily scale
    wd = 2*np.pi/48
    daily_hours = 1*(clock_hour < 36)*(clock_hour > 16)
    
    
    St = np.array([np.ones(doy.shape), t*daily_hours, np.cos(wy*doy)*daily_hours, np.sin(wy*doy)*daily_hours,
                   np.cos(wy*doy)*is_summer*daily_hours, np.sin(wy*doy)*is_summer*daily_hours, 
                   is_saturday, is_sunday, np.cos(ww*day_of_week)*daily_hours,np.sin(ww*day_of_week)*daily_hours, daily_hours,
                   #np.cos(ww*day_of_week)*np.cos(wd*clock_hour)*daily_hours, np.sin(ww*day_of_week)*np.sin(wd*clock_hour)*daily_hours,
                   np.cos(wd*clock_hour)*daily_hours, np.sin(wd*clock_hour)*daily_hours, np.cos(wd*clock_hour)*np.sin(wd*clock_hour)*daily_hours,
                   np.cos(2*wy*clock_hour), np.sin(2*wy*clock_hour),
                   np.cos(2*wd*clock_hour), np.sin(2*wd*clock_hour)]).T
    return St
```

### Trend and Seasonality Description

```python
St = Seasonality_Components_Solar(data_2014_21["t"],data_2014_21["dow"], data_2014_21["doy"], data_2014_21["SETTLEMENT_PERIOD"], data_2014_21["targetTime"])
Yt = data_2014_21['EMBEDDED_SOLAR_GENERATION'] #Target to predict

model = sm.OLS(Yt, St)
results = model.fit()
params = results.params
print("Parameters: ", results.params)
print("RMSE: ", RMSE(Yt, St@params))
```
